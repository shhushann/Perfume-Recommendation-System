```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.safari.options import Options as SafariOptions
from bs4 import BeautifulSoup
import time
import random
from urllib.parse import urljoin
import re

# --- Setup WebDriver ---
options = SafariOptions()
driver = webdriver.Safari()
driver.implicitly_wait(3)

# --- Utility Functions ---
def human_like_wait(min_sec=2, max_sec=5):
    sleep_time = random.uniform(min_sec, max_sec)

    time.sleep(sleep_time)

def simulate_mouse_movement(driver):
    try:
        body = driver.find_element(By.TAG_NAME, 'body')
        action = ActionChains(driver)
        for _ in range(random.randint(5, 10)):
            offset_x = random.randint(10, 300)
            offset_y = random.randint(10, 300)
            action.move_to_element_with_offset(body, offset_x, offset_y)
            action.perform()
            human_like_wait(0.2, 0.8)
    except Exception as e:
        print(f"Mouse movement simulation failed: {e}")

def simulate_random_click(driver):
    try:
        elements = driver.find_elements(By.CSS_SELECTOR, 'a, button')
        if elements:
            el = random.choice(elements)
            ActionChains(driver).move_to_element(el).pause(random.uniform(0.5, 1.5)).click().perform()
            print("Simulated random click.")
    except Exception as e:
        print(f"Random click simulation failed: {e}")

def simulate_scrolling(driver, scroll_times=3):
    for _ in range(scroll_times):
        scroll_direction = random.choice([1, -1])
        scroll_amount = scroll_direction * random.randint(100, 800)
        driver.execute_script(f"window.scrollBy(0, {scroll_amount});")
        simulate_mouse_movement(driver)
        if random.random() < 0.3:
            human_like_wait(2, 4)
        else:
            human_like_wait(0.5, 1.5)

def dismiss_popups(driver):
    try:
        cookie_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Accept')]")
        cookie_button.click()
        print("Dismissed popup/cookie banner.")
    except:
        pass

from urllib.parse import urljoin

def extract_perfume_links(profile_soup, section_title, base_url="https://www.fragrantica.com"):
    perfumes = []

    if section_title == "My Favorite Fragrances":
        # Handle the special container that holds both "My Favorite Fragrances" and "My Signature Fragrance"
        fav_container = profile_soup.find('div', class_="signature-favorites")
        if fav_container:
            shelf_elements = fav_container.find_all('div', class_=lambda c: c and 'shelf-element' in c)
            for cell in shelf_elements:
                link_tag = cell.find('a', href=True)
                if link_tag:
                    title = link_tag.get('title', 'No Title').strip()
                    perfume_url = urljoin(base_url, link_tag['href'])
                    perfumes.append((title, perfume_url))
    else:
        # Standard extraction logic for other sections like "Perfumes I Have", "Perfumes I Want", etc.
        header_tag = profile_soup.find(lambda tag: tag.name in ['h4', 'h5'] and tag.get_text(strip=True).lower().startswith(section_title.lower()))
        if header_tag:
            container = header_tag.find_next(lambda tag: tag.name == 'div' and tag.has_attr('class') and any('relation' in c for c in tag['class']))
            if container:
                perfume_cells = container.find_all('div', class_=lambda c: c and 'shelf-element' in c)
                for cell in perfume_cells:
                    link_tag = cell.find('a', href=True)
                    if link_tag:
                        title = link_tag.get('title', 'No Title').strip()
                        perfume_url = urljoin(base_url, link_tag['href'])
                        perfumes.append((title, perfume_url))

    return perfumes




# --- Sections to Extract ---
sections = [

    "My Favorite Fragrances",
    "Perfumes I Have",
    "Perfumes I Want",
    "Perfumes I Had"
]

profile_url = "https://www.fragrantica.com/member/2323963"
original_window = driver.current_window_handle  # Save the original tab

print(f"\n##### Visiting profile: {profile_url} #####")
try:
    driver.execute_script("window.open('');")
    driver.switch_to.window(driver.window_handles[-1])
    driver.get(profile_url)

    dismiss_popups(driver)
    human_like_wait(3, 6)
    simulate_scrolling(driver, scroll_times=random.randint(3, 6))
    simulate_random_click(driver)
    simulate_mouse_movement(driver)

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Collect all the perfume links and names
    all_perfumes = []
    for section in sections:
        perfumes = extract_perfume_links(soup, section)
        print(f"\n=== {section} ({len(perfumes)} perfumes) ===")
        for name, link in perfumes:
            print(f"- {name}: {link}")
        all_perfumes.extend(perfumes)

    # Randomly select a few
    if all_perfumes:
        num_links_to_select = random.randint(1, len(all_perfumes))
        selected = random.sample(all_perfumes, num_links_to_select)

        print(f"\n=== Selected {num_links_to_select} perfume links ===")
        for name, link in selected:
            print(f"- {name}: {link}")
    else:
        print("No perfume links found.")

    human_like_wait(2, 5)
    driver.close()
    driver.switch_to.window(original_window)

except Exception as e:
    print(f"Error visiting {profile_url}: {e}")
    if len(driver.window_handles) > 1:
        driver.close()
        driver.switch_to.window(original_window)

driver.quit()
```

```python
from urllib.parse import urlparse

def extract_brand_and_name(perfumes):
    valid_perfumes = []
    
    for title, url in perfumes:
        if title == 'No Title':  # Only process if no title is provided
            path = urlparse(url).path  # Extract URL path
            path_parts = path.split('/')[2:]  # Skip first empty string
            
            if len(path_parts) > 1:
                brand = path_parts[0]  # First part is the brand
                perfume_with_number = path_parts[1]  # Second part is the perfume name with number
                
                # Stop the name at the pattern "-number" (or digit at the end)
                perfume_name = re.split(r'(-\d+)', perfume_with_number)[0].strip()
            else:
                brand = path_parts[0]
                perfume_name = ''
            
            valid_perfumes.append((perfume_name, brand))  # Append name and brand
        else:
            # If title exists, use it as the perfume name, and we need to extract the brand from the URL
            path = urlparse(url).path
            path_parts = path.split('/')[2:]
            if len(path_parts) > 1:
                brand = path_parts[0]
                perfume_name = path_parts[1]
                
                # Stop the name at the pattern "-number" (or digit at the end)
                perfume_name = re.split(r'(-\d+)', perfume_name)[0].strip()
            else:
                brand = path_parts[0]
                perfume_name = title.strip()
            
            valid_perfumes.append((perfume_name, brand))  # Append name and brand
    
    return valid_perfumes

valid_perfumes = extract_brand_and_name(all_perfumes)
```
